# CLOSING PRICE OF STOCK

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.2, bottom=1)
sns.set_style('darkgrid')
%matplotlib inline

for i, company in enumerate(company_list, 1):
    plt.subplot(3, 2, i)
    company['Adj Close'].plot()
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {tech_list[i - 1]}")

plt.tight_layout()
```

# AVERAGE OF VARIOUS STOCKS

```python
eachday = [10, 20, 50]

for each in eachday:
    for company in company_list:
        column_name = f"AVERAGE for {each} days"
        company[column_name] = company['Adj Close'].rolling(each).mean()

fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(15)
fig.set_figwidth(20)

company_list[0][['Adj Close', 'AVERAGE for 10 days', 'AVERAGE for 20 days', 'AVERAGE for 50 days']].plot(ax=axes[0, 0])
axes[0, 0].set_title('APPLE')

company_list[1][['Adj Close', 'AVERAGE for 10 days', 'AVERAGE for 20 days', 'AVERAGE for 50 days']].plot(ax=axes[0, 1])
axes[0, 1].set_title('INTEL')

company_list[2][['Adj Close', 'AVERAGE for 10 days', 'AVERAGE for 20 days', 'AVERAGE for 50 days']].plot(ax=axes[1, 0])
axes[1, 0].set_title('TESLA')

company_list[3][['Adj Close', 'AVERAGE for 10 days', 'AVERAGE for 20 days', 'AVERAGE for 50 days']].plot(ax=axes[1, 1])
axes[1, 1].set_title('MICROSOFT')

company_list[4][['Adj Close', 'AVERAGE for 10 days', 'AVERAGE for 20 days', 'AVERAGE for 50 days']].plot(ax=axes[2, 0])
axes[2, 0].set_title('GOOGLE')

company_list[5][['Adj Close', 'AVERAGE for 10 days', 'AVERAGE for 20 days', 'AVERAGE for 50 days']].plot(ax=axes[2, 1])
axes[2, 1].set_title('AMAZON')


fig.tight_layout()
plt.show()
```

# DAILY RETURN

```python
for company in company_list:
    company['Daily Return'] = company['Adj Close'].pct_change()


fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(25)
fig.set_figwidth(27)

AAPL['Daily Return'].plot(ax=axes[0,0], legend=True, linestyle='-', marker='*')
axes[0,0].set_title('APPLE')

INTC['Daily Return'].plot(ax=axes[0,1], legend=True, linestyle='-', marker='*')
axes[0,1].set_title('INTEL')

TSLA['Daily Return'].plot(ax=axes[1,0], legend=True, linestyle='-', marker='*')
axes[1,0].set_title('TESLA')

MSFT['Daily Return'].plot(ax=axes[1,1], legend=True, linestyle='-', marker='*')
axes[1,1].set_title('MICROSOFT')

GOOG['Daily Return'].plot(ax=axes[2,0], legend=True, linestyle='-', marker='*')
axes[2,0].set_title('GOOGLE')

AMZN['Daily Return'].plot(ax=axes[2,1], legend=True, linestyle='-', marker='*')
axes[2,1].set_title('AMAZON')


fig.tight_layout()
```

```python
plt.figure(figsize=(18, 13))

for i, company in enumerate(company_list, 1):
    plt.subplot(3, 2, i)
    company['Daily Return'].hist(bins=50)
    plt.xlabel('Daily Return')
    plt.ylabel('Counts')
    plt.title(f'{company_name[i - 1]}')

plt.tight_layout()
```

```python
import yfinance as yf
import pandas as pd

tech_list = ['AAPL', 'AMZN','GOOGL','INTC','MSFT','TSLA']
end = pd.Timestamp.now()
start = end - pd.DateOffset(years=1)

closing_df = yf.download(tech_list, start=start, end=end)['Adj Close']

if 'FB' in closing_df.columns:
    tech_list.remove('FB')

closing_df = yf.download(tech_list, start=start, end=end)['Adj Close']

tech_rets = closing_df.pct_change()
tech_rets.head()

```

```python
import seaborn as sns
import pandas as pd

with pd.option_context('mode.use_inf_as_na', True):
    sns.jointplot(x='AAPL', y='GOOGL', data=tech_rets, kind='scatter', color='gold')

    sns.jointplot(x='MSFT', y='AMZN', data=tech_rets, kind='scatter', color='blue')

sns.pairplot(tech_rets, kind='reg')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')

returns_fig = sns.PairGrid(closing_df)

returns_fig.map_upper(plt.scatter, color='purple')

returns_fig.map_lower(sns.kdeplot, cmap='cool_d')

returns_fig.map_diag(plt.hist, bins=30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

custom_cmap = sns.color_palette("coolwarm", as_cmap=True)
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.heatmap(tech_rets.corr(), annot=True, cmap='summer')
plt.title('Correlation of stock return')

plt.subplot(2, 2, 2)
sns.heatmap(closing_df.corr(), annot=True, cmap='summer')
plt.title('Correlation of stock closing price')

```
